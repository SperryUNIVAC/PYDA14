```python
import pandas as pd
import statistics
import re
```

```python
data_pd_ratings = pd.read_csv('ratings.csv')
data_pd_ratings.head(10)
```

```python
# Несложно узнать, что всего пользователей у нас 610, значит пройдемся по списку их ID
userlist = [i for i in range(1, 611)]
userId_dict = {}
for i in userlist:
    usersum = len(data_pd_ratings[data_pd_ratings['userId'] == i])
    if usersum > 100:
        userId_dict.setdefault(i, usersum)
```

```python
user_1 = data_pd_ratings[data_pd_ratings['userId'] == i]
```

```python
max(user_1['timestamp'])
```

```python
min(user_1['timestamp'])
```

```python
max(user_1['timestamp']) - min(user_1['timestamp'])
```

```python
timest_delta_dict = {}
for key in userId_dict.keys():
    user_df = data_pd_ratings[data_pd_ratings['userId'] == key]
    timest_delta = max(user_df['timestamp']) - min(user_df['timestamp'])
    timest_delta_dict.setdefault(key, timest_delta)
```

```python
timest_delta_dict
```

```python
numbers = [timest_delta_dict[key] for key in timest_delta_dict]
mean_ = statistics.mean(numbers)
mean_
```
